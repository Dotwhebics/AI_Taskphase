```python
import pandas as pd
from sklearn.neural_network import MLPClassifier
```

```python
df_train = pd.read_csv('mnist_train.csv')
df_test = pd.read_csv('mnist_test.csv')
```

```python
df_train
```

```python
x_train = df_train.iloc[ : , 1 : ]
y_train = df_train.iloc[ : , 0]
```

```python
df_test
```

```python
x_test = df_test.iloc[ : , 1 : ]
y_test = df_test.iloc[ : , 0]
test_len = len(y_test)
test_len
```

```python
neural_net = MLPClassifier(activation = 'logistic', solver = 'sgd', hidden_layer_sizes = (16,16), random_state = 1)
neural_net.fit(x_train, y_train)
```

```python
predictions = neural_net.predict(x_test)
correct = 0
for i in range(test_len):
    if predictions[i] == y_test[i]:
        correct += 1
accuracy_percentage = (correct/test_len) * 100
print("The accuracy percentage is {}".format(accuracy_percentage))
```
